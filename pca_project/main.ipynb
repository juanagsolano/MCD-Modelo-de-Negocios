## PCA Analysis

```python
from sklearn.decomposition import PCA

scaler = StandardScaler(with_std=False)
scaler.fit(df)
df_scale = scaler.transform(df)
n = 2
pca = PCA(n_components=n)
pca.fit(df_scale)
```

```python
df_transform = pd.DataFrame(pca.transform(df_scale), columns=["PC"+str(i) for i in range(1,n+1)])
df_transform
```

```python
df_transform.to_csv("pca_data.csv")
```

```python
pca.transform(df_scale).shape
```

```python
variance = pca.explained_variance_
variance
```

```python
print(np.round(pca.explained_variance_ratio_*100))
```

```python
components = pd.DataFrame(pca.components_, index=["PC"+str(i) for i in range(1, n+1)], columns=df.columns)
components
```
