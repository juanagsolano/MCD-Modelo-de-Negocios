```python
import pandas as pd
import numpy as np

df = pd.read_csv("breast_cancer_wisconsin.csv")
df
```

```python
X = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(X)

df_scale = pd.DataFrame(scaler.transform(X), columns=X.columns)
df_scale
```

```python
from sklearn.decomposition import PCA

pca = PCA()
pca.fit(df_scale)
```

```python
cum_ratio = 0
ratio_list = []

for ratio in pca.explained_variance_ratio_:
    cum_ratio += ratio
    ratio_list.append(cum_ratio)
```

```python
pca_data = pd.DataFrame({"Explained Variance":pca.explained_variance_,
              "Ratio":pca.explained_variance_ratio_,
              "Cumulative": ratio_list})
pca_data
```

```python
pca_data.to_csv("explained_variance.csv")
```

```python
n = 4
pca = PCA(n_components=n)
pca.fit(df_scale)
```

```python
df_transform = pd.DataFrame(pca.transform(df_scale), columns=[
                            "PC"+str(i) for i in range(1, n+1)])
df_transform["label"] = y
df_transform
```

```python
components = pd.DataFrame(pca.components_, index=[
                          "PC"+str(i) for i in range(1, n+1)], columns=X.columns)
components
```

```python
df_transform.to_csv("pca_data.csv")
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(df_scale, y, test_size=0.2)

X_train.shape, X_test.shape
```

```python
from sklearn.ensemble import RandomForestClassifier

clf = RandomForestClassifier(n_estimators=200, min_samples_leaf=20, n_jobs=-1)

clf.fit(X_train, y_train)

y_pred = clf.predict(X_test)
```

```python
clf.feature_importances_
```

```python
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

# Create and plot confusion matrix
cm = confusion_matrix(y_test, y_pred)/len(y_test)
plt.figure(figsize=(8, 6))
ax = sns.heatmap(cm, annot=True, cmap="Blues", fmt="0.2%")
ax.set_title("Confusion Matrix - Random Forest")

ax.set_xlabel("Predicted Values")
ax.set_ylabel("True Values")

ax.xaxis.set_ticklabels(["Negativo", "Positivo"])
ax.yaxis.set_ticklabels(["Negativo", "Positivo"])
plt.show()
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

print('Metrics for best model')
print('Accuracy: {:.2f}%'.format(accuracy_score(y_test, y_pred)*100))
print('Precision: {:.2f}%'.format(precision_score(y_test, y_pred)*100))
print('Recall: {:.2f}%'.format(recall_score(y_test, y_pred)*100))
print('F1 score: {:.2f}%'.format(f1_score(y_test, y_pred)*100))
```

```python
features_importance = pd.DataFrame(
    clf.feature_importances_, index=df.iloc[:, 1:].columns).sort_values(by=0, ascending=False)
features_importance = features_importance.rename({0: "Importancia"}, axis=1)
features_importance[:20]
```
